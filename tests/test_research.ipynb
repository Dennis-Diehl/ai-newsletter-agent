```python
# Import the research node
from nodes.research import research
```

```python
# Run the research node with a minimal state
# We only pass the fields that research() actually needs
state = {
    "companies": [],
    "search_results": [],
    "existing_urls": set(),
    "raw_articles": [],
    "summaries": [],
    "newsletter": None,
    "sent": False,
}

result = research(state)
articles = result["search_results"]

print(f"Found {len(articles)} articles")
```

```python
# Inspect the results
for article in articles:
    print(f"[{article.company}] {article.title}")
    print(f"  → {article.url}")
    print()
```

```python
# Check: how many articles per company?
from collections import Counter

counts = Counter(a.company for a in articles)
for company, count in sorted(counts.items()):
    print(f"{company}: {count} article(s)")
```
